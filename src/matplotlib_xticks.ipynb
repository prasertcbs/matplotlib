```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
x=np.linspace(0, 360+1, 180) # degree
x
```

```python
x=np.linspace(0, 360+1, 180) # degree
y=np.sin(np.radians(x)) # radians
y2=np.cos(np.radians(x))
plt.figure(figsize=(14, 7))
plt.plot(x, y, label='sin', color='orange')
plt.plot(x, y2, label='cos', color='blue')
plt.legend()
plt.axhline(0, color='.7', linestyle='--')
```

```python
x=np.linspace(0, 360+1, 180)
y=np.sin(np.radians(x))
y2=np.cos(np.radians(x))
plt.figure(figsize=(14, 7))
plt.plot(x, y, label='sin')
plt.plot(x, y2, label='cos')
plt.legend(fontsize=15)
plt.xticks(np.arange(0, 360+1, 45), fontsize=20, color='red')
plt.axhline(0, color='.8', linestyle='--')
vlines=np.arange(0, 360+1, 45)
for v in vlines:
    plt.axvline(v, color='.8', linestyle='--')
```

```python
x=np.linspace(0, 360+1, 180)
y=np.sin(np.radians(x))
y2=np.cos(np.radians(x))
plt.figure(figsize=(14, 7))
plt.plot(x, y, label='sin')
plt.plot(x, y2, label='cos')
plt.legend(fontsize=20)
# plt.xticks(np.arange(0, 360+1, step=90))
plt.xticks(np.arange(0, 360+1, step=90), 
           ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'],
           fontsize=20, color='red')
plt.yticks(fontsize=15)
plt.axhline(0, color='.8', linestyle='--')
vlines=np.arange(0, 360+1, 90)
for v in vlines:
    plt.axvline(v, color='.8', linestyle='--')
```
